```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
plt.figure(figsize=(8,8))
plt.pie(Valores, labels=arreglo, autopct=make_autopct(Valores), textprops={'fontsize': 14})
plt.title("Porcentaje de Criptomonedas segÃºn su CapitalizaciÃ³n de Mercado", fontsize=20)
plt.show()
```

## Top 20 Criptomonedas segÃºn su CapitalizaciÃ³n de Mercado

```python
df2 = df.sort_values(by=["Market Capitalization"], ascending=False).nlargest(39, "Market Capitalization")
```

```python
chart = df2.plot.barh(x = 'Crypto Currency', y='Market Capitalization', figsize=(15,20), title='Top 20 Crypto monedas segÃºn capitalizaciÃ³n de mercado')
flag=True
for p in chart.patches:
    if (flag):
        chart.annotate('$ '+str(p.get_width()), (p.get_width()/1.1, p.get_y()+0.5), fontsize='large')
        flag = False
    else:
        chart.annotate('$ '+str(p.get_width()), (p.get_width()* 1.005, p.get_y()* 1.0005), fontsize='large')
```

## Conclusiones
### 1) La cantidad de crÃ­menes ocurridos y que han sido reportados en la ciudad de Seattle se concentra en el mes de enero, seguido por el mes de abril y diciembre, lo que nos indica que las temporadas crÃ­ticas para actividades de operaciÃ³n de vigilancia y prevenciÃ³n deben concentrarse en las mencionadas temporadas.

### 2) Por otro lado, la menor concentraciÃ³n de crÃ­menes reportados corresponde a los meses junio, julio y agosto; lo que nos indica una temporada de mayor seguridad en ese lapso.

### 3) Los valores mÃ¡s representativos se encuentran concentrados en el lapso de diciembre y enero, ademÃ¡s de abril, que son Ã©poca de invierno, primavera y festivos, lo que resulta coherente de la inseguirdad con esas actividades.
